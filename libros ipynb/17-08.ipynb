```python
import math

def suma(a, b):
    return a + b

def coseno(x):
    try:
        resultado = math.cos(x)
    except Exception as e:
        print(f"Error: {e}")
        resultado = None
    finally:
        print("Operación de coseno finalizada.")
    return resultado

def exponente(x):
    try:
        resultado = math.exp(x)
    except Exception as e:
        print(f"Error: {e}")
        resultado = None
    finally:
        print("Operación de exponente finalizada.")
    return resultado

def tangente(x):
    try:
        resultado = math.tan(x)
    except Exception as e:
        print(f"Error: {e}")
        resultado = None
    finally:
        print("Operación de tangente finalizada.")
    return resultado

def logaritmo_natural(x):
    try:
        resultado = math.log(x)
    except Exception as e:
        print(f"Error: {e}")
        resultado = None
    finally:
        print("Operación de logaritmo natural finalizada.")
    return resultado

def division(a, b):
    try:
        resultado = a / b
    except ZeroDivisionError:
        print("Error: No se puede dividir entre cero")
        resultado = None
    finally:
        print("Operación de división finalizada.")
    return resultado

def ejecutar_suma():
    try:
        a = float(input("Ingrese el primer número: "))
        b = float(input("Ingrese el segundo número: "))
        print("El resultado de la suma es:", suma(a, b))
    except Exception as e:
        print(f"Error: {e}")
    finally:
        print("Operación de suma finalizada.")

def ejecutar_coseno():
    try:
        x = float(input("Ingrese el número: "))
        print("El coseno de", x, "es:", coseno(x))
    except Exception as e:
        print(f"Error: {e}")
    finally:
        print("Operación de coseno finalizada.")

def ejecutar_exponente():
    try:
        x = float(input("Ingrese el número: "))
        print("El exponente de", x, "es:", exponente(x))
    except Exception as e:
        print(f"Error: {e}")
    finally:
        print("Operación de exponente finalizada.")

def ejecutar_tangente():
    try:
        x = float(input("Ingrese el número: "))
        print("La tangente de", x, "es:", tangente(x))
    except Exception as e:
        print(f"Error: {e}")
    finally:
        print("Operación de tangente finalizada.")

def ejecutar_logaritmo_natural():
    try:
        x = float(input("Ingrese el número: "))
        print("El logaritmo natural de", x,  "es:", logaritmo_natural(x))
    except Exception as e:
        print(f"Error: {e}")
    finally:
        print("Operación de logaritmo natural finalizada.")

def ejecutar_division():
    try:
        a = float(input("Ingrese el numerador: "))
        b = float(input("Ingrese el denominador: "))
        resultado = division(a, b)
        if resultado is not None:
            print("El resultado de la división es:", resultado)
    except Exception as e:
        print(f"Error: {e}")
    finally:
        print("Operación de división finalizada.")

def menu():
    while True:
        print("----- Calculadora  -----")
        print("1. Suma")
        print("2. Coseno")
        print("3. Exponente")
        print("4. Tangente")
        print("5. Logaritmo Natural")
        print("6. División")
        print("7. Salir")

        opcion = input("Seleccione una opción: ")

        if opcion == "1":
            ejecutar_suma()
        elif opcion == "2":
            ejecutar_coseno()
        elif opcion == "3":
            ejecutar_exponente()
        elif opcion == "4":
            ejecutar_tangente()
        elif opcion == "5":
            ejecutar_logaritmo_natural()
        elif opcion == "6":
            ejecutar_division()
        elif opcion == "7":
            print("Saliendo...")
            break
        else:
            print("Opción inválida. Por favor, seleccione una opción válida.")

menu()
```

```python
import datetime
"""
Constructor de la clase Persona.
Parámetros:
- nombre (str): El nombre de la persona.
- fechaNacimiento (datetime.date): La fecha de nacimiento de la persona.
"""
"""
Devuelve una representación en cadena de la persona.
Retorna:
- str: La representación en cadena de la persona.
"""
"""
Calcula la edad de la persona.
Retorna:
- int: La edad de la persona en años.
"""

class Persona:
    def __init__(self, nombre, fechaNacimiento):
        self.__nombre = nombre
        self.__fechaNacimiento = fechaNacimiento

    def __str__(self):
        return f"Nombre: {self.__nombre}, Fecha de Nacimiento: {self.__fechaNacimiento.strftime('%d/%m/%Y')}"

    def calcular_edad(self):
        fecha_actual = datetime.date.today()
        edad = fecha_actual.year - self.fechaNacimiento.year
        
        if (fecha_actual.month < self.fechaNacimiento.month or 
            (fecha_actual.month == self.fechaNacimiento.month and fecha_actual.day < self.fechaNacimiento.day)):
            edad -= 1
        return edad


persona = Persona("Jhorman", datetime.date(2002, 2, 2))


print(persona)


print(f"La edad de {persona.nombre} es: {persona.calcular_edad()} años")
```
