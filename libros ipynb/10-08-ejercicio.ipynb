2. Calcular Promedio
3. Imprimir
4. Salir

Cree un programa que capture los datos de n estudiantes:
id, Nombre completo y notas

Para las notas de poder ingresar 4 notas por estudiante

Luego de ingresados los datos se puede
a. calcular el promedio
b. Imprimir la información

Cuando se calcule el promedio todo estudiante que su promedio supere o sea igual a 8.0 debe mostrar que es 'Notable', si está entre 6.0 y 7.9 es 'Aprobado' y si es ninguno de los anteriores es 'Suspendido'

La impresión debe ser:
Por cada estudiante
ID
Nombre completo
Notas
Resultado

El Programa solo finaliza cuando selecciona la Opción salir.


```python
estudiantes = []

def ingresar_datos():
    n = int(input("Ingrese la cantidad de estudiantes: "))
    for _ in range(n):
        estudiante = {}
        estudiante['id'] = input("Ingrese el ID del estudiante: ")
        estudiante['nombre'] = input("Ingrese el nombre completo del estudiante: ")
        estudiante['notas'] = []
        for i in range(4):
            nota = float(input(f"Ingrese la nota {i+1}: "))
            estudiante['notas'].append(nota)
        estudiantes.append(estudiante)

def calcular_promedio():
    for estudiante in estudiantes:
        promedio = sum(estudiante['notas']) / 4
        estudiante['promedio'] = promedio
        if promedio >= 8.0:
            estudiante['resultado'] = 'Notable'
        elif 6.0 <= promedio <= 7.9:
            estudiante['resultado'] = 'Aprobado'
        else:
            estudiante['resultado'] = 'Suspendido'

def imprimir_informacion():
    for estudiante in estudiantes:
        print(f"ID: {estudiante['id']}")
        print(f"Nombre: {estudiante['nombre']}")
        print(f"Notas: {estudiante['notas']}")
        print(f"Promedio: {estudiante.get('promedio', 'No calculado')}")
        print(f"Resultado: {estudiante.get('resultado', 'No determinado')}")
        print("-" * 30)

def menu():
    while True:
        print("\n- - - - Menú - - - -")
        print("1. Ingresar Datos")
        print("2. Calcular Promedio")
        print("3. Imprimir")
        print("4. Salir")
        opcion = input("Seleccione una opción: ")

        if opcion == '1':
            ingresar_datos()
        elif opcion == '2':
            calcular_promedio()
        elif opcion == '3':
            imprimir_informacion()
        elif opcion == '4':
            print("Saliendo del programa...")
            break
        else:
            print("Opción no válida. Intente de nuevo.")

# Llamar al menú para ejecutar el programa
menu()
```
