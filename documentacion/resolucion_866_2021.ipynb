# Contexto General

## 1. Legislación sobre Derechos de Salud y Confidencialidad

**Ley 1751 de 2015:** Regula el derecho fundamental a la salud en Colombia, estableciendo que la intimidad y la confidencialidad de la información relacionada con la salud son derechos fundamentales. Los literales g) y k) del artículo 10 establecen la garantía de confidencialidad de toda la información suministrada en el acceso a servicios de salud y la historia clínica.

**Ley 23 de 1981:** En su artículo 34, define la historia clínica como un documento privado y confidencial que solo puede ser conocido por terceros con la autorización del paciente o en casos excepcionales previstos por la ley.

**Ley 1581 de 2012:** Establece disposiciones generales para la protección de datos personales, incluyendo los principios para el tratamiento de datos, que deben ser verídicos, completos, actualizados, y protegidos con medidas técnicas y administrativas para evitar el acceso no autorizado. Los datos de salud se consideran sensibles y requieren protección adicional.

**Ley 2015 de 2020:** Regula la historia clínica electrónica interoperable, facilitando el acceso y ejercicio de derechos relacionados con la salud, y establece que la interoperabilidad debe desarrollarse conforme a las regulaciones pertinentes.

## 2. Reglamentos y Directrices

**Decreto 620 de 2020:** Subroga ciertos títulos del Decreto 1078 de 2015, regulando aspectos relacionados con los servicios ciudadanos digitales y estableciendo lineamientos generales sobre el uso y operación de los sistemas de información.

**Resolución 1995 de 1999:** Establece normas sobre el manejo de la historia clínica y define al equipo de salud, que incluye a profesionales, técnicos, auxiliares, y auditores médicos responsables de la atención clínica y la evaluación de calidad.

**Ley 1122 de 2007 y Decreto-Ley 4107 de 2011:** Regulan el Sistema Integral de Información de la Protección Social (SISPRO) y las funciones relacionadas con los sistemas de información en el sector de salud.

**Guía de Lineamientos de los Servicios Ciudadanos Digitales - Anexo 1 (2020):** Define la interoperabilidad como la capacidad de los sistemas para intercambiar, integrar y compartir información, facilitando el cumplimiento de funciones constitucionales y legales.

# Resolución 866 de 2021

## Objetivo

La Resolución 866 de 2021 tiene como objetivo establecer normas específicas para la implementación y regulación de la historia clínica electrónica interoperable en Colombia. Su propósito es garantizar que la información de salud sea accesible, precisa y se maneje de acuerdo con las normativas de protección de datos y confidencialidad.

## Aspectos Clave

### Interoperabilidad

**Definición y Alcance:** La resolución define la interoperabilidad como la capacidad de los sistemas de información de salud para comunicarse, compartir y utilizar datos de manera efectiva entre diferentes entidades y plataformas.

**Implementación:** Establece los requerimientos técnicos y operativos para asegurar que los sistemas de historia clínica electrónica sean compatibles y puedan intercambiar información de manera segura y eficiente.

### Protección de Datos

**Confidencialidad:** Reitera la necesidad de proteger la confidencialidad de los datos de salud, alineándose con los principios establecidos en la Ley 1581 de 2012. Esto incluye asegurar que la información solo sea accesible para personas autorizadas y que se implementen medidas de seguridad adecuadas.

**Consentimiento:** La resolución refuerza que el acceso a la historia clínica electrónica debe estar basado en el consentimiento expreso e informado del titular de los datos, salvo en los casos establecidos por la ley.

### Acceso y Control

**Titularidad:** Confirma que el titular de la historia clínica electrónica es el paciente, quien tiene el control sobre el acceso y el uso de su información.

**Autorización:** Solo el titular puede autorizar a terceros para acceder a su información, excepto en situaciones donde la ley permita el acceso sin autorización previa.

### Cumplimiento y Supervisión

**Normas Técnicas:** Establece los lineamientos técnicos que deben seguir las entidades encargadas de gestionar la historia clínica electrónica para garantizar su interoperabilidad.

**Supervisión:** Define el rol de las autoridades competentes, como el Ministerio de Salud y el Ministerio de Tecnologías de la Información y las Comunicaciones, en la supervisión del cumplimiento de las normativas.

## Contextualización

La Resolución 866 de 2021 se enmarca en un contexto de modernización del sistema de salud colombiano, buscando adaptar la infraestructura y los procesos a las nuevas tecnologías y requerimientos de interoperabilidad. Con la creciente digitalización de los registros de salud, esta resolución busca asegurar que la transición hacia sistemas electrónicos se realice de manera que proteja los derechos de los pacientes, garantice la calidad de la información y facilite el acceso a servicios de salud.

En resumen, la Resolución 866 de 2021 es una pieza clave en el esfuerzo por modernizar el sistema de salud en Colombia, promoviendo una historia clínica electrónica interoperable que respete la privacidad de los pacientes y cumpla con las normativas legales.
